1. Write a Python programming to display a bar chart of the popularity of programming
Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
def bar_chart(languages,popularity):
    '''
          Description: 
                this function is displaying bar chart of the popularity og languages
          Parameters: 
                languages: list of language names
                popularity: list of popularity
          Return : 
                None
    '''
    plt.bar(languages,popularity)
    plt.title("Bar chart on pupularity of planguages")
    plt.xlabel("Programming languages")
    plt.ylabel("popularity")
    plt.show()


def main():
    programming_languages=["java","python","php","javascipt","c#","c++"]
    popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
    bar_chart(programming_languages,popularity)


if __name__=="__main__":
    main()
```

5. Write a Python programming to display a bar chart of the popularity of programming
Languages. Attach a text label above each bar displaying its popularity (float value).
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
def bar_chart_text_label(languages,popularity):
    '''
          Description: 
                this function is displaying bar chart text label 
          Parameters: 
                languages: list of language names
                popularity: list of popularity
          Return : 
                None
    '''
    colors=["red","blue","green","cyan","hotpink","orange"]
    plt.bar(languages,popularity,color=colors)
    plt.title("Bar chart on pupularity of planguages")
    plt.xlabel("Programming languages")
    plt.ylabel("popularity")
    for i in range(len(languages)):
        plt.text(i, popularity[i] + 0.5, str(popularity[i]), ha='center')

    plt.show()


def main():
    programming_languages=["java","python","php","javascipt","c#","c++"]
    popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
    bar_chart_text_label(programming_languages,popularity)


if __name__=="__main__":
    main()
```

9. Write a Python programming to display a bar chart of the popularity of programming
Languages. Increase bottom margin.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
def bar_chart_edge_color_increse(languages,popularity):
    '''
          Description: 
                this function is displaying bar chart increse the border size
          Parameters: 
                languages: list of language names
                popularity: list of popularity
          Return : 
                None
    '''
    width=[0.8,0.3,0.4,0.5,0.8,0.7]
    colors=["red","blue","green","cyan","hotpink","orange"]
    plt.bar(languages,popularity,color=colors,edgecolor="blue",width=width)
    plt.title("Bar chart on pupularity of planguages")
    plt.xlabel("Programming languages")
    plt.ylabel("popularity")
    for i in range(len(languages)):
        plt.text(i, popularity[i] + 0.5, str(popularity[i]), ha='center')
    
    plt.gcf().subplots_adjust(bottom=0.3)
    plt.xticks()
    plt.show()


def main():
    programming_languages=["java","python","php","javascipt","c#","c++"]
    popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
    bar_chart_edge_color_increse(programming_languages,popularity)


if __name__=="__main__":
    main()
```

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)

```python
def bar_chart_men_women(men_means,women_means):
    '''
          Description: 
                this function is displaying bar chart mens and womens
          Parameters: 
                men : list of items values
                women: list of item values
          Return : 
                None
    '''
    n_groups=5
    fig, ax = plt.subplots()
    index = np.arange(n_groups)
    bar_width = 0.35
    opacity = 0.8

    plt.bar(index, men_means, bar_width,alpha=opacity,color='g',label='Men')

    plt.bar(index + bar_width, women_means, bar_width,alpha=opacity,color='r',label='Women')

    plt.xlabel('Person')
    plt.ylabel('Scores')
    plt.title('Scores by person')
   
    plt.legend()
    plt.tight_layout()
    plt.show()


def main():
    means_men = [22, 30, 35, 35, 26]
    means_women = [25, 32, 30, 35, 29]  
    bar_chart_men_women(means_men,means_women) 


if __name__=="__main__":
    main()
```

11. Write a Python program to create bar plot from a DataFrame.
Sample Data Frame:
a b c d e
2 4,8,5,7,6
4 2,3,4,2,6
6 4,7,4,7,8
8 2,6,4,8,6
10 2,4,3,3,2

```python
def bar_chart_data_frame():
     '''
          Description: 
                this function is bar plot from a data frame
          Parameters: 
                axis: list of data x-axis and y-axis
          Return : 
                None
    '''
     a=np.array([[4,8,5,7,6],[2,3,4,2,6],[4,7,4,7,8],[2,6,4,8,6],[2,4,3,3,2]])
     df=pd.DataFrame(a, columns=['a','b','c','d','e'], index=[2,4,6,8,10])
     df.plot(kind="bar")
     plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
     plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
     plt.xlabel("X-axis")
     plt.ylabel("Y-axis")
     plt.title("Sample data bar on Data frame")
     plt.legend()
     plt.tight_layout()
     plt.show()


def main():

    bar_chart_data_frame()


if __name__=="__main__":
     main()

```

12. Write a Python program to create bar plots with error bars on the same figure. Sample Date
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
def error_bars(mean_valocity,standard_velocity):
    '''
          Description: 
                This function is bar plot from a data with error bars
          Parameters: 
                mean_velocity[]: list of items 
                standard_velocity[]: list of items
          Return : 
                None
    '''
    index=np.arange(len(mean_valocity))
    plt.bar(index,mean_valocity,yerr=standard_velocity,capsize=5,width=0.5,label="velocity",ecolor="r")
    plt.xlabel("x_axis")
    plt.ylabel("Y-axis")
    plt.title("Error Bar ")
    plt.show()


def main():
    mean_velocity=[0.2474, 0.1235, 0.1737, 0.1824]
    standard_velocity=[ 0.3314, 0.2278, 0.2836, 0.2645]
    error_bars(mean_velocity,standard_velocity)


if __name__=="__main__":
    main()

```

13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text
label above each bar displaying men means (integer value).
Sample Data
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
def error_bars_text_label(mean_valocity,standard_velocity):
    '''
          Description: 
                This function is bar plot from a data with error bars
          Parameters: 
                mean_velocity[]: list of items 
                standard_velocity[]: list of items
          Return : 
                None
    '''
    index=np.arange(len(mean_valocity))
    plt.bar(index,mean_valocity,yerr=standard_velocity,capsize=5,width=0.5,label="velocity",ecolor="r")
    for item in range(len(standard_velocity)):
        plt.text(item, standard_velocity[item], str(standard_velocity[item]), ha='center')
    plt.xlabel("x_axis")
    plt.ylabel("Y-axis")
    plt.title("Error Bar ")
    plt.show()


def main():
    mean_velocity=[0.2474, 0.1235, 0.1737, 0.1824]
    standard_velocity=[ 0.3314, 0.2278, 0.2836, 0.2645]
    error_bars_text_label(mean_velocity,standard_velocity)


if __name__=="__main__":
    main()

```
