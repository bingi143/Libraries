```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

1. Write a Python programming to create a pie chart of the popularity of programming
Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
def pie_chart(languages,popularity):
    '''
          Description: 
                This function is create pie chart
          Parameters: 
                languages[]: a list of items
                popularity[]: a list of items
          Return : 
                None
    '''
    plt.pie(popularity,labels=languages)
    plt.show()


def main():
    languages=["Java","python", "PHP", "JavaScript", "C#", "C++"]
    Popularity=[ 22.2, 17.6, 8.8, 8, 7.7, 6.7]
    pie_chart(languages,Popularity)


if __name__=="__main__":
    main()
```

2. Write a Python programming to create a pie chart with a title of the popularity of programming
Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
def pie_chart_title(languages,popularity):
    '''
          Description: 
                This function is create pie chart with title name
          Parameters: 
                languages[]: a list of items
                popularity[]: a list of items
          Return : 
                None
    '''
    plt.pie(popularity,labels=languages,autopct="%1.1f%%")
    plt.title("popularity of programming Languages.")
    plt.show()


def main():
    languages=["Java","python", "PHP", "JavaScript", "C#", "C++"]
    Popularity=[ 22.2, 17.6, 8.8, 8, 7.7, 6.7]
    pie_chart_title(languages,Popularity)


if __name__=="__main__":
    main()
```

3. Write a Python programming to create a pie chart of gold medal achievements of five most
successful countries in 2016 Summer Olympics. Read the data from a csv file.
Sample data:
medal.csv
country,gold_medal
United States,46
Great Britain,27
China,26
Russia,19
Germany,17

```python
def pie_chart_read_from_file(axis):
    '''
          Description: 
                This function is create pie chart with read data from file
          Parameters: 
                axis: we got data from file
          Return : 
                None
    '''
    plt.pie(axis["gold_medal"],labels=axis["country"],autopct="%1.1f%%")
    plt.title("Gold medal achivements")
    plt.show()


def main():
    axis=pd.read_csv("medal_achivement.text")
    pie_chart_read_from_file(axis)


if __name__=="__main__":
    main()
```
