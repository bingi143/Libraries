1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python
def sample_line(x_axis,y_axis):
    '''
          Description: 
                this function is draw a line with suitable label in the x axis, y axis and a title
          Parameters: 
                x_axis[]: a list of items in x_axis
                y_axis[]: a list of items in y-axis
          Return : 
                None
    '''
    plt.plot(x_axis,y_axis)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Sample line")
    plt.show()


def main():
    x_axis=[1,2,3,4,5]
    y_axis=[1,2,3,4,5]
    sample_line(x_axis,y_axis)


if __name__=="__main__":
    main()
```

3. Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1

```python
def line_plot_text_file(axis):
    '''
          Description: 
                this function is draw a line using given axis values taken from a text file
          Parameters: 
                None
          Return : 
                None
    '''
    x_axis,y_axis=axis['a'],axis['b']
    plt.plot(x_axis,y_axis)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Line Plot with Given Axis Values")
    plt.show()


def main():
    axis=pd.read_csv("text.txt",delimiter=" ",header=None,names=["a","b"])
    
    line_plot_text_file(axis)


if __name__=="__main__":
    main()
```

5. Write a Python program to plot two or more lines on same plot with suitable legends of each line.

```python
def multiple_line_chart(x_axis,y_axis_1,y_axis_2,y_axis_3):
     '''
          Description: 
                this function plot two or more lines on same plot with suitable legends of each line
          Parameters: 
                axis: list of data x-axis and y-axis
          Return : 
                None
    '''
     plt.plot(x_axis,y_axis_1,label="Line_1")
     plt.plot(x_axis,y_axis_2,label="Line_2")
     plt.plot(x_axis,y_axis_3,label="Line_3")
     plt.title("Multiple lines in same plot")
     plt.xlabel("X-axis")
     plt.ylabel("Y-axis")
     plt.legend()
     plt.show()


def main():
    x_axis=[1,2,3,4,5]
    y_axis_1=[3,4,5,6,7]
    y_axis_2=[4,5,1,2,1]
    y_axis_3=[9,4,3,2,1]
    multiple_line_chart(x_axis,y_axis_1,y_axis_2,y_axis_3)


if __name__=="__main__":
     main()

```

6. Write a Python program to plot two or more lines with legends, different widths and colors.

```python
def multiple_line_chart_width(x_axis,y_axis_1,y_axis_2,y_axis_3):
     '''
          Description: 
                this function plot two or more lines with legends different width and color
          Parameters: 
                axis: list of data x-axis and y-axis
          Return : 
                None
    '''
     plt.plot(x_axis,y_axis_1,label="Line_1",lw=3,c='r')
     plt.plot(x_axis,y_axis_2,label="Line_2",lw=4,c='b')
     plt.plot(x_axis,y_axis_3,label="Line_3",lw=2,c='g')
     plt.title("Multiple lines in same plot different width color")
     plt.xlabel("X-axis")
     plt.ylabel("Y-axis")
     plt.legend()
     plt.show()


def main():
    x_axis=[1,2,3,4,5]
    y_axis_1=[3,4,5,6,7]
    y_axis_2=[4,5,1,2,1]
    y_axis_3=[9,4,3,2,1]
    multiple_line_chart_width(x_axis,y_axis_1,y_axis_2,y_axis_3)


if __name__=="__main__":
     main()

```

9. Write a Python program to display the current axis limits values and set new axis values.

```python
def axix_limit_set(x_axis,y_axis):
    '''
          Description: 
                this function displaying the current axis limit values and setting new axis values
          Parameters: 
                x_axis: list of data x-axis values
                y_axis: list of items y-axis
          Return : 
                None
    '''
    plt.plot(x_axis,y_axis,marker="o",lw=2)
    plt.xlim(min(x_axis,),max(x_axis))
    plt.ylim(min(y_axis),max(y_axis))
    plt.xticks()
    plt.title("Current axis limit")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()
    plt.plot(x_axis,y_axis,marker="*",lw=3)
    plt.xlim([1,6])
    plt.ylim([min(y_axis),max(y_axis)])
    plt.plot(x_axis,y_axis,marker="o")
    plt.title("Setting new axis limit")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()


def main():
    x_axis=[1,2,3,4,5]
    y_axis=[5,6,5,8,9]
    axix_limit_set(x_axis,y_axis)


if __name__=="__main__":
    main()

```

11. Write a Python program to plot several lines with different format styles in one command
using arrays.

```python
def plot_different_style():
    '''
          Description: 
                this function plot of lines with different formate
          Parameters: 
                None
          Return : 
                None
    '''
    axis=np.arange(0.,5.,0.2)
    plt.plot(axis,axis,'b*',axis,axis*2,'ro',axis,axis**3,'g--',)
    plt.title("plot lines different formate")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()

def main():
    plot_different_style()


if __name__=="__main__":
    main()
```

12. Write a Python program to create multiple types of charts

```python
def multiple_chart():
    '''
          Description: 
                this function create multiple types of charts
          Parameters: 
                None
          Return : 
                None
    '''
    axis=np.arange(1,10,2)
    size=[80,30,40,90,50]
    colors=["red","green","black","yellow","blue"]
    plt.suptitle("multiple type charts")
    plt.subplot(2,2,1)
    plt.plot(axis,axis,'g--',ls="dotted")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Chart-1")
    plt.subplot(2,2,2)
    plt.title("Chart-2")
    plt.scatter(axis,axis*2,c=colors,s=size)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.subplot(2,2,3)
    plt.title("Chart-3")
    plt.bar(axis,axis*3)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.subplot(2,2,4)
    plt.title("Chart-4")
    plt.plot(axis,axis*4,c="k",ls="dashdot")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()


def main():
    multiple_chart()


if __name__=="__main__":
    main()
```
