```python
import seaborn as sbn
import pandas as pd
import matplotlib.pyplot as plt
```

1. Write a program to draw box plot of life expectancy by continent for a data set given in a
url
https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapmind
er-FiveYearData.csv

```python
def box_plot_expecatancy():
    '''
          Description: 
                this function is displaying box plot expectancy using seaborn
          Parameters: 
                None
          Return : 
                None
    '''
    axis=pd.read_csv("https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv")
    sbn.boxplot(x="lifeExp",y="continent",data=axis)
    plt.title("box plot")
    plt.xlabel("life expectancy")
    plt.ylabel("continent")
    plt.show()


def main():
    box_plot_expecatancy()


if __name__=="__main__":
    main()
```

2. Write a program to draw a box plot of day by tips for a dataset given in a url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
def box_plot_day():
    '''
          Description: 
                this function is displaying box plot day by tips
          Parameters: 
                None
          Return : 
                None
    '''
    axis=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
    sbn.boxplot(x="day",y="tip",data=axis)
    plt.title("box plot")
    plt.xlabel("Days")
    plt.ylabel("Tips")
    plt.show()


def main():
    box_plot_day()


if __name__=="__main__":
    main()
```
