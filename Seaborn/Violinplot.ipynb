```python
import seaborn as sbn
import pandas as pd
import matplotlib.pyplot as plt
```

1. Write a program to draw a violin plot of sex against total_bill for a given dataset
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
def violin_plot_seaborn():
    '''
          Description: 
                this function is displaying violen plot using seaborn
          Parameters: 
                None
          Return : 
                None
    '''
    axis=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
    sbn.violinplot(x='sex',y="total_bill",data=axis)
    plt.title("violin plot seaborn")
    plt.show()


def main():
    violin_plot_seaborn()


if __name__=="__main__":
    main()
```

2. Write a program to draw a violin plot of “species” against “sepal length” for a dataset
given in a url
https://github.com/mwaskom/seaborn-data/blob/master/iris.csv

```python
def violin_plot_seaborn_special():
    '''
          Description: 
                this function is displaying violen plot using seaborn on special lenght
          Parameters: 
                None
          Return : 
                None
    '''
    axis=pd.read_csv("https://github.com/mwaskom/seaborn-data/raw/master/iris.csv")
    sbn.violinplot(x='species',y="sepal_length",data=axis)
    plt.xlabel("species")
    plt.ylabel("sepal_length")
    plt.title("violin plot seaborn on species")
    plt.show()


def main():
    violin_plot_seaborn_special()


if __name__=="__main__":
    main()
```
