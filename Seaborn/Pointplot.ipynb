```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sbn
```

1. Write a program to draw a point plot for sex against survived for a dataset given in url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
def point_plot_seaborn():
    '''
          Description: 
                this function is displaying point chart in seaborn usign link in data
          Parameters: 
                None
          Return : 
                None
    '''
    axis=pd.read_csv("https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv")
    sbn.pointplot(x='sex',y='survived',data=axis)
    plt.title("point plot seaborn")
    plt.show()


def main():
    point_plot_seaborn()


if __name__=="__main__":
    main()

```
