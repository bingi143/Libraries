```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sbn
```

1. Write a program to draw a swarm plot of total bill against size for a given dataset
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
def swarm_plot_bills():
    '''
          Description: 
                this function is displaying swarm plot in seaborn usign link in data
          Parameters: 
                None
          Return : 
                None
    '''
    axis=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
    sbn.swarmplot(x="size",y="total_bill",data=axis)
    plt.title("Swarm plot")
    plt.xlabel("total bills")
    plt.ylabel("Sizes")
    plt.show()


def main():
    swarm_plot_bills()


if __name__=="__main__":
    main()
```
