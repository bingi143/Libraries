```python
import seaborn as sbn
import pandas as pd
import matplotlib.pyplot as plt
```

1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a
url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
def scatter_plot_seaborn():
    '''
          Description: 
                this function is displaying scatter plot in seaborn usign link in data
          Parameters: 
                None
          Return : 
                None
    '''
    axis=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
    sbn.scatterplot(x='day',y='total_bill',data=axis)
    plt.title(" scatter plot seaborn")
    plt.show()


def main():
    scatter_plot_seaborn()


if __name__=="__main__":
    main()

```
