```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sbn
```

1. Write a program to draw bar plot of sex against survived for a dataset given in the url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
def bar_plot_seaborn():
    '''
          Description: 
                this function is displaying bar chart in seaborn usign link in data
          Parameters: 
                None
          Return : 
                None
    '''
    axis=pd.read_csv("https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv")
    sbn.barplot(data=axis,x='sex',y='survived',hue="sex",palette={"male":"red","female":"green"})
    plt.title("Bar plot seaborn")
    plt.show()


def main():
    bar_plot_seaborn()


if __name__=="__main__":
    main()

```
